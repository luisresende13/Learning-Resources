# Part A

### Multiple Tables and Sub-Queries

Example of error that is avoided with sub-queries

```python
%sql SELECT * FROM EXPLOYEES WHERE SALARY > AVG(SALARY)
```

Right use

```python
%sql SELECT * FROM EMPLOYEES WHERE salary > (SELECT AVG(Salary) FROM EmPLOYEES)
```

```python
%sql SELECT F_NAME, AVG(SALARY) AS AVG_SALARY FROM EMPLOYEES
```

```python
%sql SELECT F_NAME, (SELECT AVG(SALARY) FROM EMPLOYEES) AS AVG_SALARY FROM EMPLOYEES
```

Nested SELECTS

```python
%sql SELECT f_name, (SELECT AVG(DEPT_ID_DEP) FROM DEPARTMENTS) FROM (SELECT * FROM EMPLOYEES)
```

# Part B

### Multiple Tables with Sub-Queries In WHERE Clause

```python
%sql SELECT F_NAME FROM EMPLOYEES WHERE DEP_ID IN (SELECT DEPT_ID_DEP FROM DEPARTMENTS)
```

```python
%sql SELECT * FROM EMPLOYEES WHERE DEP_ID IN (SELECT DEPT_ID_DEP FROM DEPARTMENTS WHERE LOC_ID = 'L0002')
```

```python
%sql SELECT DEPT_ID_DEP, DEP_NAME FROM DEPARTMENTS WHERE DEPT_ID_DEP IN (SELECT DEP_ID FROM EMPLOYEES WHERE SALARY > 70000)
```

Multiple Tables Names in FROM Clause: Full Joint

```python
%sql SELECT * FROM EMPLOYEES E, DEPARTMENTS D
```
