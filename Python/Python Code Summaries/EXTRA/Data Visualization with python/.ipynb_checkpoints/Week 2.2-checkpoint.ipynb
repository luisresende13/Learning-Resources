```python
# Data Preparation #1  (Timeuse)

import pandas
import numpy

timeuse=pandas.read_csv('\\Users\luisr\Data_Frames\\timeuse.csv')
timeuse.replace(':', numpy.nan, inplace=True)
timeuse[timeuse.columns.values[3:]]=timeuse[timeuse.columns.values[3:]].astype('datetime64')

for column in timeuse.columns.values[3:]:
    timeuse[column] = [i.hour+i.minute/60 if str(i)!='NaT' else numpy.nan for i in timeuse[column]] 
```

```python
# Data Preparation #2   (causes of death in france)


cod = pandas.read_csv('\\Users\luisr\Data_Frames\\causesofdeath_france.csv')

cod['Value'] = numpy.array([i.replace(' ', '') for i in cod['Value']])
cod['Value'].replace(':', numpy.nan, inplace=True)
cod['Value'] = cod['Value'].astype(float)

top_cod_list = cod[(cod['TIME']==2008) & (cod['SEX']=='Males') ].sort_values(['Value'], ascending=False, axis=0).reset_index()['ICD10'].loc[1:11].tolist()

cod_list = dict()

for i in top_cod_list:
    cod_list[i] = cod[(cod['ICD10']==i) & (cod['SEX']=='Males')]['Value'].tolist()
    
cod_list_frame = pandas.DataFrame(cod_list)
cod_list_frame.index=cod['TIME'].unique()
top_cod = cod_list_frame.copy()

top_cod.reset_index(inplace=True)
top_cod.rename(columns={'index':'TIME'}, inplace=True)
```

```python

cod_year = cod.groupby(['TIME', 'ICD10'], axis=0, as_index=False).sum()

circ_sys = cod_year[cod_year['ICD10']=='Diseases of the circulatory system (I00-I99)']

circ_sys.plot(kind='scatter', x='TIME', y='Value')

```

```python

##############    -array.plot(kind='pie', parameters...)    #############################################330
```

```python

colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink', 'brown']
explode_list = [0, 0, 0, 0, 0.07, 0.07, 0.07] # ratio for each continent with which to offset each wedge.

cod_total.iloc[1:8]['Value'].plot(kind='pie',
                            figsize=(13,5),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            #labels=None,         # turn off labels on pie chart
                            pctdistance=0.75,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            colors=colors_list,  # add custom colors
                            explode=explode_list # 'explode' lowest 3 continents
                            )

# scale the title up by 12% to match pctdistance
plt.title('Main causes of death in france', y=1.12, loc='center') 

plt.axis('equal') 

# add legend
#plt.legend(labels=cod_total.index[1:8], loc='lower left', fontsize='x-small', frameon=False) 

plt.show()
```
