## Geeting Coordenates

```python
geolocator = Nominatim(user_agent="foursquare_agent")
```

```python
coords, fails, error1, error2 = [], 0, 0, 0

for city in cities:
    #address = 'Manaus, AM'#cities[4]
    try:
        location = geolocator.geocode(city)
        coords.append([' '.join(city.split()[:-1])[:-1], location.latitude, location.longitude])
        
    except:
        try:
            location = geolocator.geocode(city)
            coords.append([' '.join(city.split()[:-1])[:-1], location.latitude, location.longitude])
            error1+=1
        except:
            try:
                location = geolocator.geocode(city)
                coords.append([' '.join(city.split()[:-1])[:-1], location.latitude, location.longitude])
                error2+=1
                            
            except: fail+=1
    if len(coords) in range(0, 900, 50): print('Appended: {}, Fail: {}, One error: {}'.format(len(coords), fails, error1))
print('Completed')
```

#    1. SEARCH VENUES ARROUND:

```python
search_query = 'School'
radius = 10000
LIMIT = 300

schools = {}
fails, error1, error2 = 0, 0, 0

for name, search, lat, lng in zip(coords_frame['Nome do Município'][:30], coords_frame['address search'][:30], coords_frame['lat'][:30], coords_frame['lng'][:30]):

    try:
        URL_venue_search = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, lat, lng, VERSION, search_query, radius, LIMIT)
        json_obj = requests.get(URL_venue_search).json()
        schools[name] = json_normalize(json_obj['response']['venues'])

    except:
        try:
            URL_venue_search = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, lat, lng, VERSION, search_query, radius, LIMIT)
            json_obj = requests.get(URL_venue_search).json()
            schools[name] = json_normalize(json_obj['response']['venues'])
            error1+=1
        except: 
            try:
                URL_venue_search = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, lat, lng, VERSION, search_query, radius, LIMIT)
                json_obj = requests.get(URL_venue_search).json()
                schools[name] = json_normalize(json_obj['response']['venues'])
                error2+=1  
            except:
                schools[name] = None
                fails+=1
                
    if len(schools) in range(0, 5, 31): print('Appended: {}, Fails: {}, error1: {}, error2: {}'.format(len(schools), fails, error1, error2))
print('\nCompleted')
                
    
```
